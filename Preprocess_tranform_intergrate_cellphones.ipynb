```python
import pandas as pd
import numpy as np
```

```python
phone_df = pd.read_csv("Data\Cellphones\Cellphones_data.csv")
phone_df.info()
```

```python
config_df = pd.read_csv("Data\Cellphones\Cellphones_specification_data.csv")
config_df.info()
```

```python
config_df.isnull().sum().sort_values()
```

```python
intergration_df = pd.DataFrame()
```

```python
intergration_df['Tên sản phẩm'] = phone_df['name']
```

```python
import re

for i in list(intergration_df.index):
    intergration_df.loc[i, ['Tên sản phẩm']] = re.split('[|I]',str(intergration_df['Tên sản phẩm'].iloc[i]))[0].split("Chính hãng")[0].strip()
```

```python
intergration_df['Giá gốc'] = phone_df['price']
intergration_df['Giá bán'] = phone_df['final_price']
intergration_df['Kích thước màn hình'] = config_df['Kích thước màn hình']
intergration_df['RAM'] = config_df['Dung lượng RAM']
intergration_df['Bộ nhớ'] = config_df['Bộ nhớ trong']
intergration_df['Chipset'] = config_df['Chipset']
intergration_df['Pin'] = config_df['Pin']
intergration_df['Kích thước'] = config_df['Kích thước']
intergration_df['Bluetooth'] = config_df['Bluetooth']
intergration_df['Camera trước'] = config_df['Camera trước']
intergration_df['Wi-Fi'] = config_df['Wi-Fi']
intergration_df['Hệ điều hành'] = config_df['Hệ điều hành']
intergration_df['Độ phân giải màn hình'] = config_df['Độ phân giải màn hình']
intergration_df['Tần số quét'] = config_df['Tần số quét']
intergration_df['Trọng lượng'] = config_df['Trọng lượng']
intergration_df['Quay video'] = config_df['Quay video']
intergration_df['GPU'] = config_df['GPU']
intergration_df['Cổng sạc'] = config_df['Cổng sạc']
intergration_df['url'] = phone_df['url']
intergration_df['image_url'] = phone_df['small_image']
```

```python
id_list = (intergration_df.query("`Giá bán` == 0").index)

intergration_df['Giá bán'].iloc[id_list] = intergration_df['Giá gốc'].iloc[id_list]

intergration_df.info()
```

```python
intergration_df.to_csv (r'Data\Cellphones\Cellphones_processed_data.csv', index=False, header=True)
```
