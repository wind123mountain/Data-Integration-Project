```python
import json
```

```python
with open(r"D:\20212\THDL\CrawlNguyenKim\CrawlNguyenKim\spiders\details.json", mode = "r", encoding="utf8") as f:
    data = json.load(f)
```

```python
columns = set()
for line in data:
    columns.update(set(line.keys()))
```

```python
df = pd.DataFrame(columns = columns)
df
```
