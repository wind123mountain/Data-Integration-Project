```python
mem_dict = {}
ram_dict = {}
size_dict = {}

mem_ram_dict = {}
mem_size_dict = {}
ram_size_dict = {}

mem_ram_size_dict = {}

# CP_df.sort_values(by=['Camera trước', 'Pin'], inplace=True)

def hash_1(name):
    dict = CP_df.groupby([name])["Tên sản phẩm"].groups

    for key in dict.keys():
        dict[key] = list(dict[key])

    return dict

mem_dict = hash_1('Bộ nhớ')
ram_dict = hash_1('RAM')
size_dict = hash_1('Kích thước màn hình')



def hash_2(name1, name2):
    index = CP_df[((CP_df[name1] == 'no_infomation') & (CP_df[name2] != 'no_infomation')) | ((CP_df[name1] != 'no_infomation') & (CP_df[name2] == 'no_infomation'))].index
    dict = CP_df[[name1, name2, "Tên sản phẩm"]].drop(index).groupby([name1, name2])["Tên sản phẩm"].groups

    for key in dict.keys():
        dict[key] = list(dict[key])

    return dict

mem_ram_dict = hash_2('Bộ nhớ', 'RAM')
mem_size_dict = hash_2('Bộ nhớ', 'Kích thước màn hình')
ram_size_dict = hash_2('RAM', 'Kích thước màn hình')



def hash_3(name1, name2, name3):
    index = CP_df[(CP_df[name1] == 'no_infomation') | (CP_df[name2] == 'no_infomation') | (CP_df[name3] == 'no_infomation')].index
    dict = CP_df.drop(index).groupby([name1, name2, name3])["Tên sản phẩm"].groups
    for key in dict.keys():
        dict[key] = list(dict[key])

    return dict

mem_ram_size_dict = hash_3('Bộ nhớ', 'RAM', 'Kích thước màn hình')

hash = {('Bộ nhớ'): mem_dict, 
        ('RAM'): ram_dict,
        ('Kích thước màn hình'): size_dict,
        ('Bộ nhớ', 'RAM'): mem_ram_dict,
        ('Bộ nhớ', 'Kích thước màn hình'): mem_size_dict,
        ('RAM', 'Kích thước màn hình'): ram_size_dict,
        ('Bộ nhớ', 'RAM', 'Kích thước màn hình'): mem_ram_size_dict}
```

### **Matching Data**

```python
from strsimpy.jaccard import Jaccard
from strsimpy.overlap_coefficient import OverlapCoefficient

overlap = OverlapCoefficient(6)
jaccard = Jaccard(3)

def matching_data(instance1, instance2):
    # if instance1['Camera trước'] != 'no_infomation':
        # if instance1['Camera trước'] < instance2['Camera trước']:
        #     return 0
        # elif instance1['Camera trước'] == instance2['Camera trước']:
            
        #     if instance1['Pin'] != 'no_infomation' and instance1['Pin'] < instance2['Pin']:
        #         return 0
        #     elif instance1['Pin'] != 'no_infomation' and instance1['Pin'] > instance2['Pin']:
        #         return -1

        # elif instance1['Camera trước'] > instance2['Camera trước']:
        #     return -1

    if instance1['Camera trước'] != 'no_infomation' and instance2['Camera trước'] != 'no_infomation':
        if instance1['Camera trước'] != instance2['Camera trước']:
            return 0
    if instance1['Pin'] != 'no_infomation' and instance2['Pin'] != 'no_infomation':
        if instance1['Pin'] != instance2['Pin']:
            return 0
    

    sim_name = jaccard.similarity(instance1['Tên sản phẩm'], instance2['Tên sản phẩm'])

    same_plus = instance1['Tên sản phẩm'].find('PLUS')*instance2['Tên sản phẩm'].find('PLUS')
    same_pro = instance1['Tên sản phẩm'].find('PRO')*instance2['Tên sản phẩm'].find('PRO')
    same_max = instance1['Tên sản phẩm'].find('MAX')*instance2['Tên sản phẩm'].find('MAX')

    rate = 1
    if (same_max < 0 or same_plus < 0 or same_pro < 0):
        rate = 0.01

    if instance1['Chipset'] != 'no_infomation' and instance2['Chipset'] != 'no_infomation':
        sim_chipset = overlap.similarity(instance1['Chipset'], instance2['Chipset'])
        score = 0.7*sim_name + 0.3*(sim_chipset) # 0.7, 0.3
    else:
        score = sim_name
        
    if score*rate > 0.7: #0.65
        return 1
    
    return 0
```

```python
def update_hash(key_hash, key, index):
    if type(key) is tuple:
        if key in hash[key_hash]:
            hash[key_hash][key].append(index)
        else:
            hash[key_hash][key] = [index]

        if len(key) == 3:
            if (key[0], key[1]) in mem_ram_dict:
                mem_ram_dict[(key[0], key[1])].append(index)
            else:
                mem_ram_dict[(key[0], key[1])] = [index]

            if (key[0], key[2]) in mem_size_dict:
                mem_size_dict[(key[0], key[2])].append(index)
            else:
                mem_size_dict[(key[0], key[2])] = [index]

            if (key[1], key[2]) in ram_size_dict:
                ram_size_dict[(key[1], key[2])].append(index)
            else:
                ram_size_dict[(key[1], key[2])] = [index]
        
        if len(key) > 1:
            for i in range(len(key)):
                if key[i] in hash[key_hash[i]]:
                    hash[key_hash[i]][key[i]].append(index)
                else:
                    hash[key_hash[i]][key[i]] = [index]

    else:
        if key in hash[key_hash]:
            hash[key_hash][key].append(index)
        else:
            hash[key_hash][key] = [index]

    return 0
```

```python
def matching(input_df):
    for i in list(input_df.index):
        key_hash = ['Bộ nhớ', 'RAM', 'Kích thước màn hình']
        key = []

        if input_df['Bộ nhớ'].loc[i] == 'no_infomation':
            key_hash.remove('Bộ nhớ')
        else:
            key.append(input_df['Bộ nhớ'].loc[i])

        if input_df['RAM'].loc[i] == 'no_infomation':
            key_hash.remove('RAM')
        else:
            key.append(input_df['RAM'].loc[i])

        if input_df['Kích thước màn hình'].loc[i] == 'no_infomation':
            key_hash.remove('Kích thước màn hình')
        else:
            key.append(input_df['Kích thước màn hình'].loc[i])
        

        if len(key_hash) > 0:
            if len(key_hash) == 1:
                key_hash = key_hash[0]
                key = key[0]
            else:
                key_hash = tuple(key_hash)
                key = tuple(key)

            if key in hash[key_hash]:
                id_list = hash[key_hash][key]
            else:
                if len(key_hash) == 2:
                    id_list = hash[key_hash][('no_infomation', 'no_infomation')]
                if len(key_hash) == 1:
                    id_list = hash[key_hash][('no_infomation', 'no_infomation')]
                if len(key_hash) == 3:
                    id_list = [206]

            isMatching = False
            for id in id_list:
                res = matching_data(input_df.loc[i], df.loc[id])
                if res == 1:
                    map_dict[id].append((input_df.loc[i][-1], i))
                    isMatching = True
                    break
                # elif res == -1:
                #     break

            if not isMatching:
                index = len(df.index)
                
                df.loc[index] = input_df.loc[i]
                update_hash(key_hash, key, index)
                map_dict[index] = [(input_df.loc[i][-1], i)]
        
        else:
            except_instance.append((input_df.loc[i][-1], i))
```

```python
import re

no_matching_id_worldphone = []
matching_id_worldphone = []

def matching_worldphone(input_df):

    df_sort_name = df.sort_values(by=['Tên sản phẩm'])
    id_list = df_sort_name.index

    for i in list(input_df.index):
        isMatching = False

        score_list = []
        
        type1 = input_df.loc[i]['Tên sản phẩm'].split()[0]
        for id in id_list:
            type2 = df_sort_name.loc[id]['Tên sản phẩm'].split()[0]
            # print(type1 + " " + type2)

            if type1 > type2:
                continue
            elif type1 == type2:
                storage = re.findall(r'[\d+.0|\d+]+[G|T]+B', input_df.loc[i]['Tên sản phẩm'])

                if len(storage) > 0 and df_sort_name.loc[id]['Bộ nhớ'] != 'no_infomation':
                    if storage[0] != df_sort_name.loc[id]['Bộ nhớ']:
                        continue
                ip_name = ' '.join(filter(None, re.split(r'[\d+.0|\d+]+[G|T]+B| ', str(input_df['Tên sản phẩm'].iloc[i]))))
                sim_name = 4*overlap.similarity(ip_name, df_sort_name.loc[id]['Tên sản phẩm']) + jaccard.similarity(ip_name, df_sort_name.loc[id]['Tên sản phẩm'])

                if sim_name/5 > 0.95:
                    # map_dict[id].append((input_df.loc[i][-1], i))
                    isMatching = True
                    score_list.append((id, sim_name))
                    # break

            else:
                break

        if not isMatching:
            # index = len(df.index)
            # map_dict[index] = [(input_df.loc[i][-1], i)]
            no_matching_id_worldphone.append(i)

        else:
            score_list.sort(key=lambda tup: tup[1], reverse=True)
            # map_dict[score_list[0][0]].append((input_df.loc[i][-1], i))
            matching_id_worldphone.append((i, score_list[0][0], score_list[0][1]))
            # matching_id_worldphone.append(((i, input_df.loc[i]['Tên sản phẩm']), (score_list[0][0], df_sort_name.loc[score_list[0][0]]['Tên sản phẩm'])))


matching_worldphone(tranWP_df)
   
```
